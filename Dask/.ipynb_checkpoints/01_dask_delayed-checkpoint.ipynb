```python
from time import sleep

def inc(x):
    sleep(1)
    return x + 1

def add(x, y):
    sleep(1)
    return x + y
```

```python
%%time

x = inc(1)
y = inc(2)
z = add(x, y)
```

## Parallelize with the dask.delayed decorator

```python
from dask import delayed
```

```python
%%time

x = delayed(inc)(1)
y = delayed(inc)(2)
z = delayed(add)(x, y)
```

```python
%%time

z.compute()
```

```python
z
```

```python
z.visualize()
```

## Parallelize a for-loop

```python
data = [1,2,3,4,5,6,7,8]
```

```python
%%time

results = []
for i in data:
    y = inc(x)
    results.append(y)
    
total = sum(results)
```

```python
%%time

results = []

for i in data:
    y = delayed(inc)(x)
    results.append(y)
    
total = delayed(sum)(results)
print("Before computing:", total)  # Let's see what type of thing total is
result = total.compute()
print("After computing :", result)  # After it's computed
```

```python
total.visualize()
```

## Parallelize a for-loop with control flow

```python
def double(x):
    sleep(1)
    return 2 * x

def is_even(x):
    return not x % 2

data = [i for i in range(1,11)]
```

```python
%%time

results = []
for i in data:
    if is_even(i):
        y = double(i)
    else:
        y = inc(i)
    results.append(y)
    
total = sum(results)
print(total)
```

```python
%%time

results = []
for i in data:
    if is_even(i):
        y = delayed(double)(i)
    else:
        y = delayed(inc)(i)
    results.append(y)
    
total = delayed(sum)(results)
```

```python
%time total.compute()
```

```python
total.visualize()
```

```python
from glob import glob
filenames = sorted(glob(os.path.join('data', 'nycflights', '*.csv')))
```

```python
%%time

sums = []
counts = []

for fn in filenames:
    df = pd.read_csv(fn)
    
    by_origin = df.groupby('Origin')
    
    total = by_origin.DepDelay.sum()
    
    count = by_origin.DepDelay.count()
    
    sums.append(total)
    counts.append(count)
    
total_delays = sum(sums)
n_flights = sum(counts)
mean = total_delays / n_flights
```

```python
mean
```

```python
from dask import compute
```

```python
%%time

sums = []
counts = []

for fn in filenames:
    df = delayed(pd.read_csv)(fn)    # delayed
    
    by_origin = df.groupby('Origin')
    
    total = by_origin.DepDelay.sum()
    
    count = by_origin.DepDelay.count()
    
    sums.append(total)
    counts.append(count)
    
sums, counts = compute(sums, counts)  # computed

total_delays = sum(sums)
n_flights = sum(counts)
mean = total_delays / n_flights
```
